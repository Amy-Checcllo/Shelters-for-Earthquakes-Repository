```python
import pandas as pd
import geopandas as gpd
```

## Shapefiles de Manzanas

```python
# Lee archivo shapefile de distritos del norte de Irlanda.
# Antes hay que asegurarse de crear la carpeta Shapefiles y subir todos los archivos de Shapefiles.zip
gdf_districts = gpd.read_file('MANZANA.shp') # Distritos del norte de Irlanda

# muestra el tipo de dato
type(gdf_districts)
```

```python
gdf_Lima = gdf_districts.copy()
gdf_Lima = gdf_Lima[(gdf_Lima.DEPARTAMEN == 'LIMA') & (gdf_Lima.PROVINCIA == 'LIMA')]
```

```python
gdf_Lima.head(5)
```

```python
gdf_Lima.shape
```

### Agregación de latitud y longitud

```python
gdf_Lima_coor = gdf_Lima.copy()

# Calcular los centroides si las geometrías no son puntos
gdf_Lima_coor['centroid'] = gdf_Lima_coor.geometry.centroid

# Extraer latitud y longitud desde los centroides
gdf_Lima_coor['latitud'] = gdf_Lima_coor['centroid'].y
gdf_Lima_coor['longitud'] = gdf_Lima_coor['centroid'].x
```

```python
# Eliminar la columna 'centroid' ya que no puede haber 2 columnas de tipo geometry
gdf_Lima_coor = gdf_Lima_coor.drop(columns=['centroid'])
```

```python
gdf_Lima_coor['GEOMETRY'] = gdf_Lima_coor.geometry
# Eliminar múltiples columnas que no guardan información relevante
gdf_Lima_coor = gdf_Lima_coor.drop(columns=['geometry'])
```

```python
gdf_Lima_coor.head(3)
```

## Base de datos con Manzanas caracterizadas

```python
df_manzanas_carac = pd.read_excel('BD_ManzanasCarac.xlsx')
```

```python
df_manzanas_carac
```

```python
df_manzanas_carac.shape
```

```python
# Cambiar el nombre de cobertura de Agua y Electricidad
df_manzanas_carac = df_manzanas_carac.rename(columns={'COB_AP17': 'COB_AGUA','COB_EE17': 'COB_ELEC'})
```

```python
# Convertir la columna ID_MANZANA a string
df_manzanas_carac['ID_MANZANA'] = df_manzanas_carac['ID_MANZANA'].astype(str)
```

```python
df_manzanas_carac = df_manzanas_carac.drop(columns=['NOMBDIST_VULN','TIPO_EE17'])
```

## Union de Manzanas caracterizadas con shapefile

```python
# CRUZAR DATOS DE SHAPEFILE
# Realizar el inner join entre df_manzanas_carac y gdf_combined_coor
# La clave de unión es "ID_MANZANA"
gdf_manzanas_carac = df_manzanas_carac.merge(gdf_Lima_coor, left_on='ID_MANZANA', right_on='IDMANZANA', how='inner')
```

```python
gdf_manzanas_carac = gdf_manzanas_carac.drop(columns=['IDMANZANA','CODDPTO','CODPROV','CODDIST','CODZONA','SUFZONA','CODMZNA','SUFMZNA','CODCCPP','NOMCCCPP','IDCCPP'])
```

```python
gdf_manzanas_carac.head(3)
```

```python
# Checar las columnas con nombre de mas de 10 caracteres
for col in list(gdf_manzanas_carac.columns):
    if len(col)>10:
        print(col,len(col))
```

```python
# Descargar el archivo de los albergues potenciales como SHAPEFILE
gdf_manzanas_carac = gpd.GeoDataFrame(gdf_manzanas_carac, geometry='GEOMETRY')
```

```python
# Exportar a shapefile
gdf_manzanas_carac.to_file('manzanas_caracterizadas_lima.shp', driver='ESRI Shapefile')
```
